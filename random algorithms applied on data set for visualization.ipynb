```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Load data from CSV file
data = pd.read_csv('houseRent.csv')

# Remove rows with missing values
data = data.dropna()

# Encode categorical variable to numerical
le = LabelEncoder()
data['type'] = le.fit_transform(data['type'])

# Scale or normalize the data
scaler = StandardScaler()
data[['price', 'sqfeet', 'beds', 'baths']] = scaler.fit_transform(data[['price', 'sqfeet', 'beds', 'baths']])

```

```python
sns.histplot(data=data.head(2), x='beds')
sns.scatterplot(data=data.head(10), x='sqfeet', y='price')
corr_matrix = data.head(10).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python

# Feature selection
X = data[['sqfeet', 'beds', 'baths', 'type']]
y = data['price']
selector = SelectKBest(score_func=f_regression, k=2)
X_new = selector.fit_transform(X, y)
scores = selector.scores_
print(scores)
```

```python

# Model selection and training
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv('houseRent.csv')

# Create a histogram of the price attribute
plt.hist(data['price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Rental Prices')
plt.show()

# Create a scatter plot of the sqfeet and price attributes
plt.scatter(data['sqfeet'], data['price'])
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.title('Relationship between Square Feet and Price')
plt.show()

# Create a bar chart of the average rental price by type
data.groupby('type')['price'].mean().plot(kind='bar')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.title('Average Rental Price by Type')
plt.show()
```
